```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
questions = pd.read_csv('dataedit.csv')
quizzes   = pd.read_csv('quiz.csv')
test1 = questions.loc[questions['quiz_id'] == 41]
test1["update_at"] = test1["update_at"].astype("datetime64")
test1.update_at.groupby([test1["update_at"].dt.day, test1["update_at"].dt.hour]).count().plot(kind="bar")
plt.xlabel("den hodina")
plt.xticks(rotation=60, fontsize=7)
plt.ylabel("# studentu")
plt.show()
```
